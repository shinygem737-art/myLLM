```python
import math

def min_operations(a, b):
    """
    计算最少操作次数，每次操作可将任意多个元素同时乘以2。
    假设一定有解，且数组元素为正整数。
    """
    max_k = 0
    for ai, bi in zip(a, b):
        if bi % ai != 0:
            # 无法通过乘以2的幂得到，无解
            return -1
        q = bi // ai
        # 检查 q 是否为 2 的幂
        #if q & (q - 1) != 0:
        #   return -1
        if not math.log2(q).is_integer():
            return -1
        # 计算 k = log2(q)
        #k = q.bit_length() - 1
        k = int(math.log2(q))
        max_k = max(max_k, k)
    return max_k

a = [1, 1, 2, 4]
b = [1, 4, 4, 4]
c = [1, 3, 9]
d = [2, 6, 9]
min_operations(a, b)
#min_operations(c, d)
```
